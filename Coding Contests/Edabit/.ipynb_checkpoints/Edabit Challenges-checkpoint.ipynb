## Find the perimeter of a Rectangle

[Source](https://edabit.com/challenge/Yx2a9B57vXRuPevGh)

Create a function that takes `height` and `width` and finds the perimeter of a rectange.

**Examples**

(6, 7) => 26

(20, 10) => 60

(2, 9) => 22

```python
# Given the height and width, return the perimeter of a rectange
def perimeter(height, width):
    return (2*height) + (2*width)

print(perimeter(6, 7))
print(perimeter(20, 10))
print(perimeter(2, 9))
```

## How Many Vowels?

[Source](https://edabit.com/challenge/p88k8yHRPTMPt4bBo)

Create a function that takes a string and returns the number (count) of vowels contained within the string.

**Examples**

`count_vowels("Celebration") => 5
 count_vowels("Palm") => 1
 count_vowels("Prediction") => 4`
 
**Notes**

* a, e, i, o, u are considered vowels (not y)

* All test cases are one word and only contain letters

```python
def count_vowels(text):
    vowels = ["a", "A", "e", "E", "i", "I", "o", "O", "u", "U"]
    count = 0
    for letter in text:
        if letter in vowels:
            count = count + 1
        else:
            pass
    return count
    
print(count_vowels("Hello"))
print(count_vowels("Pineapple"))
```

## Alphabet Soup

[Source](https://edabit.com/challenge/4Agr8CTY7x2rAhh5n)

Create a function that takes a string and returns a string with its letters in alphabetical order.

**Examples**

`alphabet_soup("hello") => "ehllo"
 alphabet_soup("edabit") => "abdeit"
 alphabet_soup("hacker") => "acehkr"`

**Notes**

You can assume numbers and punctuation symbols won't be included. Inputs are single words.

```python
# Turn string into character list
# Sort the list
# Turn list into string using the join method

def alphabet_soup(text):
    lst = tuple(text)
    lst = sorted(lst)
    string = "".join(lst)
    return string

print(alphabet_soup("hello"))
print(alphabet_soup("geek"))
```

## Factorial of an Integer

[Source](https://edabit.com/challenge/FF6kYPHdAcJnoosr5)

Create a function that takes an integer and returns the factorial of that integer.

**Examples**

`factorial(3) => 6
 factorial(5) => 120
 factorial(13) => 6227020800`
 
**Note**

Assume all inputs are greater than or equal to 0.

```python
def factorial(N):
    if N <= 1:
        return 1
    else:
        return N * factorial(N-1)

print(factorial(3))
print(factorial(5))
```

## Compound Interest

[Source](https://edabit.com/challenge/HjpihSKFBfRCCg86J)

Suppose that you invest $10,000$ for 10 years at an interest rate of $6%$ compounded monthly. What will be the value of your investment at the end of the 10 year period?

Create a function that accepts the principal `p`, the term in years `t`, the interest rate `r`, and the number of compounding periods per year `n`. The function returns the value at the end of term rounded to the nearest cent.

**Examples**

`compound_interest(100, 1, 0.05, 1) => 105.0

 compound_interest(3500, 15, 0.1, 4) => 15399.26
 
 compound_interest(100000, 20, 0.15, 365) => 2007316.26
`

**Notes**

The interest rate is given as a decimal

## Return the Volume of a Cone

[Source](https://edabit.com/challenge/uXPF7XM4KoiiSnjuA)

Create a function that takes the height and radius of a cone as arguments and returns the volume of the cone.



**Examples**

volCone(3, 2) => 12.57

volCone(18, 0) => 0

**Notes**

* Return approximate answer by rounding the answer to the nearest hundredth.

* Use Python's `math.pi` constant

* If the cone has no volume, return $0$

```python
# Volume of a cone is 1/3*pi*r^2*h

def coneVol(height, radius):
    import math
    return round((1/3*math.pi*radius**2*height),2)

print(coneVol(3, 2))
print(coneVol(15, 6))
print(coneVol(18, 0))
```

## Stand in Line

[Source](https://edabit.com/challenge/S26tvW7BPrJsyJApt)

Write a function that takes a list and a number as arguments. Add the number to the end of the list, then remove the first element of the list. The function should then return the update list.

**Examples**

`next_in_line([5, 6, 7, 8, 9], 1)  [6, 7, 8, 9, 1]

next_in_line([7, 6, 3, 23, 17], 10)  [6, 3, 23, 17, 10]

next_in_line([1, 10, 20, 42 ], 6)  [10, 20, 42, 6]

next_in_line([], 6)  "No list has been selected"`

**Notes**

For an empty list input, return: "No list has been selected"

```python
# List methods needed:
    # pop: remove an item by index
    # append: add item to end of list
def next_in_line(lst: list, num: int):
    if len(lst) != 0:
        lst.append(num) # Add the number to the end of the list
        lst.pop(0) # Remove the first element of a list
        return lst
    else:
        return "No list has been selected"

print(next_in_line([5, 6, 7, 8, 9], 1))
print(next_in_line([], 6))   
```

## Recursion: Sum

[Source](https://edabit.com/challenge/si2H6WC5YX99cn6LQ)

Write a function that finds the sum of the first `n` natural numbers. Make sure your function contains a recursive element.

**Examples**

`sum_numbers(5) -> 15
 sum_numbers(1) -> 1
 sum_numbers(12) -> 78
 `

```python
# Just like the factorial, only addition this time

def sum_numbers(n):
    if n <= 1:
        return 1
    else:
        return n + sum_numbers(n - 1)
    

print(sum_numbers(5))
print(sum_numbers(12))
```

## Lowercase and Uppercase Map

[Source](https://edabit.com/challenge/pimnBHXJNtQffq4Cf)

Write a function that creates a dictionary with each (**key, value**) pair being the (**lower case, upper case**) versions of a letter, respectively.

**Examples**

`mapping(["p", "s"]) => {"p": "P", "s": "S"}`

**Notes**

All of the letters in the input list will always be lowercase.

```python
# return each letter and its uppercase counterpart
# Put results into a dictionary
def mapping(letters: list):
    results = {} # Start with an empty dictionary
    for i in letters: # For each letter in the list
        results[i] = i.upper() # Return the uppercase letter
    return results

mapping(["p", "s"])
```

## Find the Missing Number

[Source](https://edabit.com/challenge/oMCNzA4DcgpsnXTRJ)

Create a function that takes a list of numbers between 1 and 10 (excluding 1 number) and returns the missing number.

**Examples**

`missing_num([1, 2, 3, 4, 6, 7, 8, 9, 10]) => 5
 missing_num([7, 2, 3, 6, 5, 9, 1, 4, 8]) => 10
 missing_num([10, 5, 1, 2, 4, 6, 8, 3, 9]) => 7
`

**Notes**

* The list of numbers will be unsorted

* Only one number will be missing

```python
# Given a list of integers between 1 and N, find the one missing number in the sequence
# Sum of whole list = n(n+1)/2 = number theory formula
# Find sum of actual list
# Subtract to find the missing number

def missing_num(lst: list) -> int:
    N = max(lst)
    theory = (N * (N + 1)) / 2
    actual = sum(lst)
    MissingNum = theory - actual
    return int(MissingNum)

missing_num([1, 2, 3, 5])
missing_num([1,2,3,4,6,7,8,9,10])
```

## Apocalyptic Numbers

[Source](https://edabit.com/challenge/8MiJwz7fdaWRiRDYc)

A number `n` is **apocalyptic** if $2^{n}$ contains a string of 3 consecutive 6's (666 being the presumptive "number of the beast").

Create a function that takes a number `n` as input. If the number is apocalyptic, find the index of $666$ in $2^{n}$ and return `"Repent! X days until the Apocalypse!"` (X being the index). If not, return `"Crisis averted. Resume sinning."`.

**Examples**

`

apocalyptic(109) => "Crisis averted. Resume sinning."

apocalyptic(157) => "Repent! 9 days until the Apocalypse!"

apocalyptic(175) => "Crisis averted. Resume sinning."

apocalyptic(220) => "Repent! 6 days until the Apocalypse!"

`

```python
def apocalyptic(N):
    num = str(2 ** N) # Calculate the number 2^n
    if "666" in num:  # Check for 666 substring
        position = num.find("666") # Return the position of the substring
        print(f"Repent! {position} days until the Apocalypse!")
    else: # Otherwise, if it's not True
        print("Crisis averted. Resume sinning")


apocalyptic(109)
apocalyptic(157)
apocalyptic(175)
apocalyptic(220)
```

## Find the Mean of All Digits

[Source](https://edabit.com/challenge/BZ4mMcEz3aqosEtbC)

Create a function that returns the mean of all digits.

**Examples**

`
mean(42) => 3
mean(666) => 6
mean(12345) => 3
`

**Notes**

* The mean of all digits is the sum of the digits / how many digits there are

* The mean will always be an integer

```python
def meanDigits(num):
    string = str(num)
    strings = list(string)
    integers = list(map(int, strings))
    return sum(integers) / len(integers)

print(meanDigits(666))
print(meanDigits(12345))
```

## The Karaca's Encryption Algorithm

## Calculate the Profit

[Source](https://edabit.com/challenge/YfoKQWNeYETb9PYpw)

You work for a manufacturer, and have been asked to calculate the total profit made on the sales of a product. You are given a dictionary containing the *cost price per unit* (in dollars), *sell price per unit* (in dollars), and the starting inventory. Return the total **profit** made, rounded to the nearest dollar. Assume all of the inventory has been sold.

**Examples**

`
profit({
    "cost_price": 32.67,
    "sell_price": 45.00,
    "inventory": 1200}) => 14796
`

**Notes**

Profit = Total Sales - Total Cost

```python
def profit(info: dict):
    cost, sell, amount = info["cost_price"], info["sell_price"], info["inventory"]
    return round(((sell*amount) - (cost*amount)), 0)


info = {
  "cost_price": 32.67,
  "sell_price": 45.00,
  "inventory": 1200
       }

profit(info)
```

## Filter out Strings from a List

[Source](https://edabit.com/challenge/EfEpbcGjXQYDFcdxF)

Create a function that takes a list of strings and integers, and filters out the list so that it returns a list of integers only.

**Examples**

`
filter_list([1, 2, 3, "a", "b", 4]) => [1, 2, 3, 4]
filter_list(["A", 0, "Edabit", 1729, "Python", "1729"]) => [0, 1729]
filter_list(["Nothing", "Here"]) => []
`

**Notes**

Don't overthink this one.

```python
def filter_list(lst: list)->list:
    return [i for i in lst if type(i) is int]

list1 = [1, 2, 3, "a", "b", 4]

print(filter_list(list1))
print(filter_list(["Nothing", "Here", "There", "Is"]))
```

## Even Number Generator

[Source](https://edabit.com/challenge/eJLwXpuaffjFnzENN)

Using list comprehensions, create a function that finds all even numbers from `1` to the given number.

**Examples**

`
find_even_nums(8) => [2, 4, 6, 8]
find_even_nums(4) => [2, 4]
find_even_nums(2) => [2]
`

**Notes**

* Try to use list comprehensions instead of logic
* If there are no even numbers, return an empty list

```python
def find_even_nums(num: int):
    nums = list(range(1, num+1))
    return [i for i in nums if i%2 == 0]

find_even_nums(8)
```
