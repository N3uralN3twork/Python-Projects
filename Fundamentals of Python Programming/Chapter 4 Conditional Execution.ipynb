### 4.1 Boolean Expressions <a class="anchor" id="4.1 Boolean Expressions"></a>

Can only have one of two values: True or False

Comes from the British mathematician George Boole

```python
print(False, type(False))
a = True
x = 10
y = 9
```

### 4.3 The Simple if Statement



* "if" is a reserved statement

* The condition is a boolean expression that determines whether the code will be executed

* The block is one or more statements that will run if the condition holds True

How many spaces should you indent? 

Answer: at least one, but Python automatically does it for you

You can also just press the TAB key to indent yourself

**BEWARE:** You need to use the same indent distance consistently for your program to run, best practice is to use **spaces** over **tab** for some reason.

### 4.4 The if/else Statement

The **else** block contains an alternate block of code that the program executes when the condition is False.



* "if" begins the if/else statement

* "else" is a reserved statement and a : must follow it

* the else-block are statements to be executed if the above condition is False

```python
if x == 2:
    print(x)
else:
    print("X does not equal 2")
```

### 4.5 Compound Boolean Expressions

A combination of simple Boolean expressions

* Equal: ==

* And: &

* Or: |

* Not: !

```python
if x and y == 20:
    print("X and Y equal 20")
else:
    print("X and Y not equal to 20")
```

### 4.6 The pass Statement

Sometimes, you can write illegal if-else statements

Like "else" statements requiring some code block instead of being empty

Luckily, there is something called a "pass" statement which means *do nothing*

Can be useful as a placeholder for future code

```python
if x == 10:
    pass # Do nothing
else:
    print("X is not 9")
```

### 4.7 Floating-point Equality

== checks for *exact* equality

Basically checking to see if 2 floating-point numbers are close enough to each other.

```python
diff1 = 1.11 - 1.10
diff2 = 2.11 - 2.10
diff = diff1 - diff2
if diff < 0:
    diff = -diff
if diff < 0.000001:
    print(f"Same, diff={diff}")
else:
    print("Different")
```

### 4.8 Nested Conditionals

Statements within if statements can also be other if statements

Like the example below:

```python
value = int(input("Please enter an integer in the range of 0 to 10"))
if value >= 0:
    if value <= 10:
        print("Within Range")
else:
    pass # Do nothing remember
```

### 4.9 Multi-way Decision Statements

The introduction of **elif** statement

The general form of an **if/elif/else** statement:



**BEWARE**: An **if/elif/else** statement that omits the **else** statement may fail to execute properly if none of the conditions are True.

```python
x = 10
if x == 10:
    print("X is 10")
elif x == 1:
    print("X is 1")
else:
    pass
```
