# Chapter 9 Objects

So far, we've been using functions, but there exists a more powerful technique that uses *objects*.

Python is an *object-oriented* language, which allows the user to define, create, and manipulate objects. Objects bundle together data and functions. Like other variables, each Python object has a type/class.

A *class* is like an object constructor, or a "blueprint" for creating objects.

This chapter explores some of the classes available in Python.

### 9.1 Using Objects

Some examples of objects that we've used so far include integers, floating-point numbers, strings, and functions.

A typical object contains two things, *data* and *methods*.

*Data* consists of an object's instance variables, attributes, or fields

*Methods* are like functions, also known as operations.

*Members* are the methods and instance variables of an object

### 9.2 String Objects

```python
name = input("Please enter your name: ")
print(f"Hello {name.upper()}, how are you?")
```

`name.upper()` is a method call.

The general form of a method call is as follows:


* *object* is an expression that represents an object (name)

* The period, which is called *dot*, associates an object expression with the method to be called.

* *methodname* is the name of the method to be run

* The *parameterlist* is a comma-separated list of parameters to the method. Sometimes, this list may be empty.

Method calls work like functions and can return a value to its caller. Methods may also accepts parameters.

### 9.3 File Objects

Most useful applications require greater data persistance, like a word processor that allows you to save any changes you've made to a document. Text files store character data, and you can use an editor to make changes to any text file.

Python has a file class that makes it easy to make objects that can store data to, and retrieve from, the disk.

The statement:

`f = open("myfile.txt", "r")`

creates and returns a file object named `f`. The first argument to open is the name of the file, and the second argument is a mode. The following modes are supported:

* "r" opens the file for reading

* "w" opens the file for writing

* "a" opens the file to append data to it

It is always important to close a file after opening it. Thankfully, Python offers a way to express code that automatically closes the file when finished via the `with/as` statement.
